```python
X = data["House Size (sq.ft.)"]
```

```python
Y = data["House Price"]
```

```python
plt.scatter(X,Y)
plt.show()
```

```python
plt.scatter(X,Y)
plt.axis([0, 2500, 0, 1500000])
plt.xlabel("House Price (sq.ft.)")
plt.ylabel("House Price")
plt.show()
```

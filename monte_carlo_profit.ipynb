```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
rev_m = 170
rev_stdev = 20
iterations = 1000
```

```python
rev = np.random.normal(rev_m, rev_stdev, iterations)
rev
```

```python
plt.figure(figsize=(15,6))
plt.plot(rev)
plt.show()
```

```python
np.random.normal(0.6, 0.1)
```

```python
# assume cost of goods sold is around 60% of revenues, with 10% standard deviation
COGS = - (rev * np.random.normal(0.6, 0.1))

plt.figure(figsize=(15, 6))
plt.plot(COGS)
plt.show()
```

```python
COGS.mean()
```

```python
COGS.std()
```

```python
Gross_Profit = rev + COGS
Gross_Profit

plt.figure(figsize=(15,6))
plt.plot(Gross_Profit)
plt.show()
```
